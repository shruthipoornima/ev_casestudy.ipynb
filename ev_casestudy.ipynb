```python
from scipy.stats import chi2_contingency
```

```python
def chi_square_test(col4, col5):
    table = pd.crosstab(udf[col4], udf[col5])
    chi2, p, dof, expected = chi2_contingency(table)
    print(f"\nCross-tab between {col4} and {col5}")
    print(table)
    print(f"chi2 = {chi2: .3f}, p-value = {p:3f}")
    if p < 0.05:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    return table, pd.DataFrame(expected, index=table.index, columns=table.columns)
```

```python
chi_square_test("fuel_used", "travel_cost (per month)")
chi_square_test("choices_on_fuel_hike", "fuel_used")
chi_square_test("aware_ev_scheme_FAME II", "interest_in_getting_electricvehicle")
chi_square_test("if_govt_gave_50%_subsidy_would_you_switch_to_EV?", "interest_in_getting_electricvehicle")
chi_square_test("fuel_used", "private_vehicles_causing_pollution")
```

```python
chi_square_test("city", "transport_used")
chi_square_test("features_of_public_transport_in_your_location", "adopting_improved_public_transport")
chi_square_test("value_transport_scale (1=cost-effectiveness to 5-comfortability", "adopting_improved_public_transport")
```

```python
chi_results = {}

chi_tests = [
    ("fuel_used", "travel_cost (per month)"),
    ("choices_on_fuel_hike", "fuel_used"),
    ("aware_ev_scheme_FAME II", "interest_in_getting_electricvehicle"),
    ("if_govt_gave_50%_subsidy_would_you_switch_to_EV?", "interest_in_getting_electricvehicle"),
    ("fuel_used", "private_vehicles_causing_pollution"),
    ("city", "transport_used"),
    ("features_of_public_transport_in_your_location", "adopting_improved_public_transport"),
    ("value_transport_scale (1=cost-effectiveness to 5-comfortability", "adopting_improved_public_transport")
]


for col4, col5 in chi_tests:
    table, expected_df = chi_square_test(col4, col5)
    chi2, p, dof, expected = chi2_contingency(table)

    chi_results[f'{col4} vs {col5}'] = {
        'chi2': chi2,
        'p_value': p,
        'dof': dof,
        'observed': table.to_json(),  # Store as JSON string
        'expected': expected_df.to_json() # Store as JSON string
    }

chi_results_df = pd.DataFrame.from_dict(chi_results, orient='index')
file_path = 'example.xlsx'

with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    chi_results_df.to_excel(writer, sheet_name='Chi_Square_Results')
    # You can add more sheets here for observed/expected tables if needed
    # For example:
    # for key, value in chi_results.items():
    #     observed_df = pd.read_json(value['observed'])
    #     expected_df = pd.read_json(value['expected'])
    #     observed_df.to_excel(writer, sheet_name=f'{key}_Observed')
    #     expected_df.to_excel(writer, sheet_name=f'{key}_Expected')
```
